```python
# Configurar variable de entorno para evitar advertencia de MKL
import os
os.environ["OMP_NUM_THREADS"] = "1"

# Librer√≠as necesarias
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Seleccionar caracter√≠sticas relevantes
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
data_selected = data[features]

# Escalar los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)

# Ajustar modelo K-Means
optimal_k = 5  # N√∫mero de clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Agregar los clusters al DataFrame original
data_encoded = data.copy()
data_encoded['Cluster'] = clusters

print(data_encoded.head())
```

```python
# Seleccionar solo las columnas num√©ricas para evitar problemas con cadenas de texto
numeric_features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']

# Calcular la media de cada caracter√≠stica num√©rica por cluster
cluster_means = data_encoded.groupby('Cluster')[numeric_features].mean()

# Graficar heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cluster_means, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Media de Caracter√≠sticas por Cluster')
plt.xlabel('Caracter√≠sticas')
plt.ylabel('Cluster')
plt.show()
```

```python
# Reducir dimensiones con PCA
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data_scaled)

# Crear un DataFrame para visualizaci√≥n
data_2d_df = pd.DataFrame(data_2d, columns=['PCA1', 'PCA2'])
data_2d_df['Cluster'] = clusters

# Graficar en 2D
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='PCA1', y='PCA2', hue='Cluster', palette='tab10', data=data_2d_df, alpha=0.7
)
plt.title('Clusters Visualizados en 2D (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster', loc='best')
plt.show()
```
